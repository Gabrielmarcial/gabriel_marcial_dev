# 🧠 **Estruturas Condicionais em Python**


### 👉 O que é a instrução `if ... else`?

Em programação, nem sempre queremos que todo o código seja executado da mesma forma. Às vezes, queremos **tomar decisões** com base em certas condições.  
É aí que entra a estrutura condicional `if`.

Ela permite que você **execute um bloco de código apenas se uma condição for verdadeira**. Se a condição for falsa, você pode optar por executar um **bloco alternativo** com `else`.

---

### 🧱 **Estrutura básica:**


    if condição:
        # bloco de código executado se a condição for verdadeira
    else:
        # bloco de código executado se a condição for falsa

````python
num1 = int(input())
num2 = int(input())

if num1>num2:
  print('Num1 é maior que Num2.')

print('oi')
````

### 🔄 O que é a instrução `elif`?

Às vezes, você quer testar **várias condições diferentes**, não apenas uma com `if` e uma alternativa com `else`.  
Nesses casos, usamos o `elif` (abreviação de "**else if**").

A instrução `elif` permite verificar **outras condições** se a primeira não for satisfeita.  
Você pode usar **quantos `elif` quiser**, e o Python vai testar **de cima pra baixo**, parando na **primeira condição verdadeira**.

---

### 🧱 **Estrutura com `elif`:**

```python
if condição_1:
    # executado se condição_1 for verdadeira
elif condição_2:
    # executado se condição_1 for falsa e condição_2 for verdadeira
elif condição_3:
    # executado se as anteriores forem falsas e condição_3 for verdadeira
else:
    # executado se nenhuma condição acima for verdadeira


````python
num1 = int(input())
num2 = int(input())

if num1>num2:
  print('Num1 é maior que Num2.')

elif num1==num2:
  print('São iguais')

else:
  print('nenhuma das alternativas')
````

### ⚡ Usando `if...else` em uma linha (expressão condicional)

Em Python, é possível escrever uma condição `if...else` de forma **mais curta e direta** usando o que chamamos de **expressão condicional** ou **expressão ternária**.

---

### 🧱 **Sintaxe:**

```python
valor_se_verdadeiro if condição else valor_se_falso


````python
num1 = int(input())
num2 = int(input())

result = 0

if num1>num2:
  result  = num1 + num2
else :
  result  = num1 * num2

print(result)
````

````python
num1 = int(input())
num2 = int(input())

result = num1 + num2 if num1>num2 else num1*num2

print(result)
````
