## 🔁 **Laço `for` em Python**

O `for` é uma das estruturas de repetição mais utilizadas em Python.  
Ele permite **percorrer uma sequência de itens** (como listas, strings ou números) e **executar um bloco de código para cada item**.

---

### 🧱 **Sintaxe do `for`:**

```python
for variavel in sequencia:
    # código a ser executado

## 🔢 A função `range()`

A função `range()` é muito usada com `for` para gerar sequências numéricas de forma eficiente.

---

### 🧱 **Sintaxe:**

```python
range(stop)                 # vai de 0 até stop - 1
range(start, stop)          # vai de start até stop - 1
range(start, stop, passo)   # define o incremento (positivo ou negativo)


## 🔁 **Laço `while` em Python**

O loop `while` é usado para **executar repetidamente um bloco de código** enquanto uma **condição for verdadeira**.

É ideal para situações em que **não sabemos previamente quantas vezes o código deve repetir**, e a repetição depende de uma **condição dinâmica**.

---

### 🧱 **Sintaxe do `while`:**

```python
while condição:
    # código a ser executado


## 🔄 **Instruções `break` e `continue` em Python**

Em Python, as instruções `break` e `continue` são usadas para **alterar o fluxo de execução** dentro de loops (`for` ou `while`).

Normalmente, um loop continua até que sua condição de parada seja falsa, mas com `break` e `continue`, podemos **interromper ou pular partes da repetição** de forma controlada.

---



### 🛑 **`break`**

A instrução `break` **interrompe o loop completamente**, mesmo que a condição ainda seja verdadeira.

### 🔁 Instrução `continue` em Python

A instrução `continue` é usada dentro de loops (`for` ou `while`) para **pular a iteração atual** e ir direto para a **próxima repetição** do loop.

Ou seja, quando o Python encontra um `continue`, ele **ignora o restante do código** dentro do loop naquela iteração e **volta para o topo** do laço.

---
