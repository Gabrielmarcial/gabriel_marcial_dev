# 🔁 **Entrada e Saída de Dados em Python**


### 👉 O que são entrada e saída de dados?

Em um programa, **entrada de dados** é quando o usuário fornece informações ao sistema. Já a **saída de dados** é quando o sistema mostra uma informação de volta ao usuário.

Em Python, usamos a função `input()` para **receber dados do usuário**, e a função `print()` para **mostrar informações na tela**.

---

### 🧱 **Funções principais:**

- `input()` → Lê uma informação digitada pelo usuário (sempre retorna uma **string**).
- `print()` → Exibe mensagens ou variáveis no console.

---

````python
nome = input('Entre com seu nome:')
idade = int(input('Entre com a sua idade'))
````

````python
print('Seu nome é : ')
print(nome)
print('Sua Idade é ')
print(idade)
````

````python
print('Seu nome é : ',nome,'Sua Idade é',idade)
````

````python
print(f'Seu nome é {nome}, e Sua idade é {idade}')
````

````python
print(f'Seu nome é {nome}\nSua idade é {idade}')
````

````python
print(f'''
-------------------
Seu nome é {nome}
-------------------
Sua idade é {idade}
-------------------
''')
````

### 👉 O que são tipos de dados?

Todo valor em Python pertence a um **tipo de dado**. Isso ajuda o Python a entender **como tratar e armazenar** cada informação.

Você pode descobrir o tipo de uma variável ou valor usando a função `type()`.

---

### 🔍 Exemplo com `type()`:

```python
print(type(10))         # <class 'int'>
print(type("Python"))   # <class 'str'>
print(type(3.14))       # <class 'float'>
print(type(True))       # <class 'bool'>
```
---

````python
idade = 28
type(idade)
````

````python
print(type(10))
````

````python
nome = input('Entre com seu nome:')
type(nome)
````

### 🔢 **Números: `int()` ou `float()`**

- `int` → Números inteiros (sem casas decimais)  
  **Exemplos:** `2`, `4`, `5`, `56`

- `float` → Números decimais (com ponto flutuante)  
  **Exemplos:** `2.5`, `4.5`, `5.7`

```python
idade = 30        # int
altura = 1.75     # float
```

````python
float(30)
````

````python
int(1.75)
````

````python
idade = float(input('Entre com a sua idade:'))
idade
````

````python
idade = input('Entre com a sua idade:')
print(type(idade))
idade = float(idade)
print(type(idade))
````

### 📌 **Resumo dos principais tipos:**

| Tipo      | Função     | Exemplos               |
|-----------|------------|------------------------|
| Inteiro   | `int()`    | `2`, `10`, `50`        |
| Decimal   | `float()`  | `2.5`, `3.14`, `7.0`    |

| Booleano  | `bool()`   | `True`, `False`        |

# ➕ **Operações em Python**




### 🧮 **Operadores Aritméticos**

Os operadores aritméticos são usados para realizar operações matemáticas com variáveis e valores.

| Operador | Descrição                                                        | Exemplo        |
|----------|------------------------------------------------------------------|----------------|
| `+`      | Adição - Soma dois operandos ou mais                             | `x + y`        |
| `-`      | Subtração - Subtrai o operando direito do esquerdo               | `x - y`        |
| `*`      | Multiplicação - Multiplica dois operandos                        | `x * y`        |
| `/`      | Divisão - Divide o operando esquerdo pelo direito (float)        | `x / y`        |
| `%`      | Módulo - Resto da divisão                                        | `x % y`        |
| `//`     | Divisão inteira (piso)                                           | `x // y`       |
| `**`     | Exponenciação - Potência                                         | `x ** y`       |

---

Esses operadores são a base para a construção de expressões, condições e cálculos em Python!


````python
10 + 10
````

````python
'10' + '10' # concatena
````

````python
'10' + str(10)
````

### 🧾 **Operadores Comparativos (Relacionais)**

Usados para comparar valores. Sempre retornam `True` ou `False`.

| Operador | Descrição                                            | Exemplo        |
|----------|------------------------------------------------------|----------------|
| `>`      | Maior que                                            | `x > y`        |
| `<`      | Menor que                                            | `x < y`        |
| `==`     | Igual a                                              | `x == y`       |
| `!=`     | Diferente de                                         | `x != y`       |
| `>=`     | Maior ou igual a                                     | `x >= y`       |
| `<=`     | Menor ou igual a                                     | `x <= y`       |


````python
comparacao = 10 > 5
````

````python
type(comparacao)
````

````python
10 == 5
````

````python
'G' == 'g'
````

````python
ord('g')
````

````python
ord('G')
````

````python
chr(71)
````

### 🔗 **Operadores Lógicos**

Usados para combinar expressões booleanas.

| Operador | Descrição                                     | Exemplo                    |
|----------|-----------------------------------------------|----------------------------|
| `and`    | Retorna `True` se **ambas** as expressões forem verdadeiras | `x > 5 and y < 10` |
| `or`     | Retorna `True` se **uma ou ambas** forem verdadeiras        | `x > 5 or y < 10`  |
| `not`    | Inverte o valor lógico                         | `not True` → `False`       |


````python
10 > 5 and 3==5
````

````python
10 > 5 or 3==3
````

````python
not 10 > 11
````

````python
11 >= 11
````
